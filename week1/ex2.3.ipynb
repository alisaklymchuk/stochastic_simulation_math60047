```python
import numpy as np
import matplotlib.pyplot as plt
rng = np.random.default_rng(42)
```

### Ex. 2.3: Algorithm for sampling a Weibull random variable from Uniform[0,1]

First, we sample U ~ Unif[0,1].

Then, we turn them into X ~ Exp(1) using the inverse CDF:
$$
X_i = - \log(1-U_i)
$$

Finally, we use a transformation $$ W_i = X_i^{1/\beta} $$

The resulting W is going to be Weibull distributed.

```python
def weibull_dist(beta, n):
    unif = rng.uniform(0, 1, n)
    exp = -np.log(1-unif)
    w = exp**(1/beta)
    return w
```

```python
beta = 2.0
n = 10000
x = np.arange(0, n, 1)
weibull = rng.weibull(beta, n)
my_weibull = weibull_dist(beta, n)
```

```python
plt.figure()
plt.hist(weibull, bins=100, density=True)
plt.title("True Weibull Distribution")
plt.show()
plt.figure()
plt.hist(my_weibull, bins=100, density=True)
plt.title("Weibull Distribution from Uniform")
plt.show()
```
