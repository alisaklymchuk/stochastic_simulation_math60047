```python
import numpy as np
import matplotlib.pyplot as plt
rng = np.random.default_rng(108)
```

```python
w1 = 0.2993
w2 = 0.7007

A1 = np.array([[0.4, -0.3733], [0.06, 0.6]])
A2 = np.array([[-0.8, -0.1867], [0.1371, 0.8]])

b1 = np.array([0.3533, 0.0]).T
b2 = np.array([1.1, 0.1]).T

f1 = lambda x: A1 @ x + b1
f2 = lambda x: A2 @ x + b2

N = 100000
x = np.zeros(2).T
x_arr = [x]
for n in range(1, N):
    r = rng.uniform(0, 1)
    x = f1(x) if r < w1 else f2(x)
    x_arr.append(x)

x_arr = np.array(x_arr)
plt.scatter(x_arr[20:N, 0], x_arr[20:N, 1], s=0.1, color=[0.8, 0, 0])

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)

ax.set_xticks([])
ax.set_yticks([])

ax.set_xlim(0, 1.05)
ax.set_ylim(0, 1)

plt.show()

```
