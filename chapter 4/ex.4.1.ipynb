```python
import numpy as np
import matplotlib.pyplot as plt
rng = np.random.default_rng(108)
```

```python
w1 = 0.2993
w2 = 0.7007

A1 = np.array([[0.4, -0.3733], [0.06, 0.6]])
A2 = np.array([[-0.8, -0.1867], [0.1371, 0.8]])

b1 = np.array([0.3533, 0.0]).T
b2 = np.array([1.1, 0.1]).T

f1 = lambda x: A1 @ x + b1
f2 = lambda x: A2 @ x + b2

N = 1000
x = np.zeros(2).T

for n in range(N):
    bern = rng.binomial(n=1, p=0.5)
    i = w1 if bern else w2
    x = f1(i) if bern else f2(i)
```
